```python
# Import Libraries
import pandas as pd
import folium
```

```python
# Load the data
csv_file = 'Grand_Rapids_Art_geocodio_a54ea9b029afcf639e028b883a803ac1133476ce.csv'
df = pd.read_csv(csv_file)
df = df[['Longitude', 'Latitude', 'Property Owner','Address 1', 'ZIP code', 'Sponsor', 'Type of art', 'Description', 'Date Signed']]
```

```python
# Loading the map
gr_art = folium.Map(location=[42.96, -85.67], zoom_start=10, tiles='OpenStreetMap')
gr_art
```

```python
# Adding Popup with information
for index, row in df.iterrows():
    text = f"""
    <h3><b>{row['Property Owner']}</b></h3><b3>
    <p style="text-align:left;"> Address: <b>{row['Address 1']}</b></p>
    <p style="text-align:left;"> Art type: <b>{row['Type of art']}</b></p>
    <p style="text-align:left;"> Description: <b>{row['Description']}</b></p>
    <p style="text-align:left;"> Date signed: <b>{row['Date Signed']}</b></p>
    </p>
    """
    folium.Marker(
        [row['Latitude'], row['Longitude']],
        popup=folium.Popup(text, max_width=400),
        tooltip='Click here for more information',
        icon=folium.Icon(color="pink", icon_color="white", icon='paint-brush', prefix='fa')
        ).add_to(gr_art)

gr_art
```
